```python
import requests

API_URL = ""

def search_report(key_word):
    query_url = f"{API_URL}/api?key_word={key_word}"
    response = requests.get(query_url)
    return response.json()
```

```python
import pandas as pd

json = search_report("人工知能")
df = pd.read_json(json)
df['report'].head()
```
